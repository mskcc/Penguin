```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Define the directory where the files are located
directory = "/juno/cmo/bergerlab/sumans/Project_ecDNA/Production/penguin/data/projects/ECS_Integration_Tests.7500n/output/merged"  # Replace with your actual directory name

# Define file names
file1 = "merged.ECHO_results.csv"
file2 = "merged_ecDNA_results_filtered_p_ecDNA.tsv"

# Construct full file paths using os.path.join
file1_path = os.path.join(directory, file1)
file2_path = os.path.join(directory, file2)
```

```python
# Load the files using the constructed paths
# old_data = pd.read_csv(file1_path, header=None, names=["sample_id_concat", "chromosome", "gene", "log2", "cn", "ecDNA_prob", "ecDNA_status"])
old_data = pd.read_csv(file1_path)
new_data = pd.read_csv(file2_path, sep="\t")
```

```python
# Extract sample ID by splitting on the first three hyphens
old_data['sample_id'] = old_data['sample_id'].apply(lambda x: "-".join(x.split("-")[:4]))
```

```python
# Filter the old data where ecDNA_prob is between 0.4 and 0.5, and status is "non-ecDNA"
filtered_old_data = old_data[(old_data['ecDNA_prob'] >= 0.4) & 
                             (old_data['ecDNA_prob'] < 0.5) & 
                             (old_data['ecDNA_status'] == "non-ecDNA")]
```

```python
# Merge the filtered old data with the new data on sample_id and gene
comparison_df = pd.merge(filtered_old_data, new_data, on=["sample_id", "gene"], suffixes=('_old', '_new'))
```

```python
comparison_df.head(6)
```

```python
# Find cases in the new version where ecDNA_prob is >= 0.5 and status is "ecDNA"
rescued_cases = comparison_df[(comparison_df['ecDNA_prob_new'] >= 0.5) & 
                              (comparison_df['ecDNA_status_new'] == "ecDNA")]
```

```python
# Display the final results
rescued_cases[['sample_id', 'gene', 'ecDNA_prob_old', 'ecDNA_prob_new', 'ecDNA_status_old', 'ecDNA_status_new']]
```

```python
# Print lengths of all dataframes
print("Length of original old_data:", len(old_data))
print("Length of filtered_old_data:", len(filtered_old_data))
print("Length of original new_data:", len(new_data))
print("Length of comparison_df:", len(comparison_df))
print("Length of rescued_cases:", len(rescued_cases))
```

```python
import matplotlib.pyplot as plt

# Count the number of rescued cases for each gene
gene_counts = rescued_cases['gene'].value_counts()
total_rescued = gene_counts.sum()  # Calculate the total count of rescued cases

# Calculate the percentage of rescued cases
total_filtered_cases = len(filtered_old_data)  # Total cases in the filtered old data
percentage_rescued = (total_rescued / total_filtered_cases) * 100  # Calculate the percentage

# Plot the results
plt.figure(figsize=(12, 6))
ax = gene_counts.plot(kind='bar', color="skyblue")
plt.title("Number of Rescued Cases per Gene")
plt.xlabel("Gene")
plt.ylabel("Number of Rescued Cases")
plt.xticks(rotation=45, ha='right')

# Increase y-axis limit to ensure labels fit within the plot
plt.ylim(0, gene_counts.max() * 1.3)  # Adds 30% padding above the tallest bar

# Add text labels on top of each bar
for i, count in enumerate(gene_counts):
    plt.text(i, count + (gene_counts.max() * 0.02), str(count), ha='center', va='bottom', fontsize=10)

# Add total count and percentage annotation inside the plot
plt.text(len(gene_counts) - 1, gene_counts.max() * 1.25,
         f"Total Rescued Cases: {total_rescued} ({percentage_rescued:.1f}% of {total_filtered_cases})",
         ha='right', fontsize=12, fontweight='bold')

plt.show()
```

```python
# Filter for ecDNA+ cases in both files
ecDNA_plus_old = old_data[old_data['ecDNA_status'] == "ecDNA"]
ecDNA_plus_new = new_data[new_data['ecDNA_status'] == "ecDNA"]

# Merge to find cases that are ecDNA+ in both files
ecDNA_consistent = pd.merge(ecDNA_plus_old, ecDNA_plus_new, on=["sample_id", "gene"], suffixes=('_old', '_new'))

# Count the number of consistently ecDNA+ cases per gene
consistent_gene_counts = ecDNA_consistent['gene'].value_counts()
```

```python
total_old_ecDNA_plus = len(ecDNA_plus_old)
total_new_ecDNA_plus = len(ecDNA_plus_new)
total_consistent_ecDNA_plus = len(ecDNA_consistent)
percentage_old = (total_consistent_ecDNA_plus / total_old_ecDNA_plus) * 100
percentage_new = (total_consistent_ecDNA_plus / total_new_ecDNA_plus) * 100
```

```python
# Print lengths of all dataframes
print("Length of old_data:", len(old_data))
print("Length of new_data:", len(new_data))
print("Length of ecDNA_plus_old:", len(ecDNA_plus_old))
print("Length of ecDNA_plus_new:", len(ecDNA_plus_new))
print("Length of ecDNA_consistent:", len(ecDNA_consistent))
```

```python
# Plot the results
plt.figure(figsize=(12, 6))
ax = consistent_gene_counts.plot(kind='bar', color="skyblue")
plt.title("Number of Consistent ecDNA+ Cases per Gene")
plt.xlabel("Gene")
plt.ylabel("Number of Consistent ecDNA+ Cases")
plt.xticks(rotation=45, ha='right', fontsize=6)  # Reduce font size of gene names

# Add text labels on top of each bar with reduced font size
for i, count in enumerate(consistent_gene_counts):
    plt.text(i, count + 0.5, str(count), ha='center', va='bottom', fontsize=6)  # Smaller label size

# Add a text box with the percentage and counts information
textstr = (f"Total Old ecDNA+ Cases: {total_old_ecDNA_plus}\n"
           f"Total New ecDNA+ Cases: {total_new_ecDNA_plus}\n"
           f"Consistent ecDNA+ Cases: {total_consistent_ecDNA_plus}\n"
           f"Percentage of Old Cases Consistent: {percentage_old:.1f}%\n"
           f"Percentage of New Cases Consistent: {percentage_new:.1f}%")

# Place the text box within plot boundaries
props = dict(boxstyle='round', facecolor='white', alpha=0.75)
plt.gcf().text(0.82, 0.65, textstr, fontsize=9, va='top', ha='right', bbox=props)

# Display the plot
plt.show()
```

```python
# Filter for cases that were previously non-ecDNA or not called in the old file
non_ecDNA_old = old_data[old_data['ecDNA_status'] == "non-ecDNA"]
new_ecDNA_plus = new_data[new_data['ecDNA_status'] == "ecDNA"]

# Merge to find new cases now called ecDNA+ but were non-ecDNA or missing in the old file
non_to_ecDNA = pd.merge(new_ecDNA_plus, non_ecDNA_old, on=["sample_id", "gene"], how="left", suffixes=('_new', '_old'))
newly_rescued_cases = non_to_ecDNA[non_to_ecDNA['ecDNA_status_old'].isna() | (non_to_ecDNA['ecDNA_status_old'] == "non-ecDNA")]

# Count the number of newly called ecDNA+ cases per gene
newly_rescued_gene_counts = newly_rescued_cases['gene'].value_counts()

import matplotlib.pyplot as plt

# Assume `newly_rescued_gene_counts` is already calculated and sorted
# Select the top 20 genes with the most rescued cases
top_genes = newly_rescued_gene_counts.head(20)

# Plot the results with adjusted settings
plt.figure(figsize=(15, 6))  # Increase figure width for readability
ax = top_genes.plot(kind='bar', color="skyblue")
plt.title("Top 20 Genes with Newly Rescued ecDNA+ Cases")
plt.xlabel("Gene")
plt.ylabel("Number of Newly Rescued ecDNA+ Cases")

# Rotate x-axis labels vertically
plt.xticks(rotation=90, ha='center', fontsize=8)

# Add text labels on top of each bar
for i, count in enumerate(top_genes):
    plt.text(i, count + 0.5, str(count), ha='center', va='bottom', fontsize=8)

plt.tight_layout()  # Adjust layout to fit everything neatly
plt.show()
```

```python
# Filter ecDNA+ cases and select p_ecDNA scores
old_ecDNA_scores = old_data[old_data['ecDNA_status'] == "ecDNA"]['ecDNA_prob']
new_ecDNA_scores = new_data[new_data['ecDNA_status'] == "ecDNA"]['ecDNA_prob']

import matplotlib.pyplot as plt
import numpy as np

# Define bins
bins = np.linspace(0.5, 1.0, 20)  # Adjust as needed to match your data range and bin size

# Plot histograms for each version side-by-side
plt.figure(figsize=(10, 6))
plt.hist(old_ecDNA_scores, bins=bins, alpha=0.8, label="Old Version", color="salmon", edgecolor="black", histtype="bar", width=0.02)
plt.hist(new_ecDNA_scores, bins=bins, alpha=0.8, label="New Version", color="royalblue", edgecolor="black", histtype="bar", width=0.02)

# Labels and title
plt.title("Distribution of p_ecDNA Scores for ecDNA+ Cases (Side-by-Side Histogram)")
plt.xlabel("p_ecDNA Score")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Option 2: Density Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(old_ecDNA_scores, shade=True, label="Old Version", color="skyblue")
sns.kdeplot(new_ecDNA_scores, shade=True, label="New Version", color="lightcoral")
plt.title("Density of p_ecDNA Scores for ecDNA+ Cases")
plt.xlabel("p_ecDNA Score")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python

# Filter ecDNA+ cases strictly by p_ecDNA >= 0.5 in each version
old_ecDNA_scores = old_data[(old_data['ecDNA_status'] == "ecDNA") & (old_data['ecDNA_prob'] >= 0.5)]['ecDNA_prob']
new_ecDNA_scores = new_data[(new_data['ecDNA_status'] == "ecDNA") & (new_data['ecDNA_prob'] >= 0.5)]['ecDNA_prob']

# Combine data into a single DataFrame for easier plotting
scores_df = pd.DataFrame({
    "p_ecDNA": pd.concat([old_ecDNA_scores, new_ecDNA_scores]),
    "Version": ["Old"] * len(old_ecDNA_scores) + ["New"] * len(new_ecDNA_scores)
})

# Option 1: Violin Plot with Boxplot Overlay
plt.figure(figsize=(10, 6))
sns.violinplot(x="Version", y="p_ecDNA", data=scores_df, inner=None, palette={"Old": "skyblue", "New": "lightcoral"})
sns.boxplot(x="Version", y="p_ecDNA", data=scores_df, width=0.15, showcaps=True,
            notch=True, meanline=True, showmeans=True, meanprops={"color": "black", "linestyle": "--"})
plt.title("Distribution of p_ecDNA Scores with CI for ecDNA+ Cases")
plt.xlabel("Version")
plt.ylabel("p_ecDNA Score")
plt.show()

# Option 2: Swarm Plot with Error Bars
plt.figure(figsize=(10, 6))
sns.swarmplot(x="Version", y="p_ecDNA", data=scores_df, palette={"Old": "skyblue", "New": "lightcoral"}, alpha=0.6)
mean_std_dev = scores_df.groupby("Version")["p_ecDNA"].agg(['mean', 'std'])

# Add error bars to show Â±1 standard deviation
for i, version in enumerate(mean_std_dev.index):
    mean = mean_std_dev.loc[version, 'mean']
    std_dev = mean_std_dev.loc[version, 'std']
    plt.errorbar(i, mean, yerr=std_dev, fmt='o', color='black', capsize=5)

plt.title("Swarm Plot of p_ecDNA Scores with Error Bars (Â±1 SD) for ecDNA+ Cases")
plt.xlabel("Version")
plt.ylabel("p_ecDNA Score")
plt.show()
```
